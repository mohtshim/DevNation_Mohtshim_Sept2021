# Question no 2

```python
def count_difference(df,energy,GDP,ScimEn):
    temp_df = pd.merge(energy,GDP,how = 'inner',on = 'Country')
    data = pd.merge(temp_df,ScimEn,how = 'outer',on = 'Country')
    temp_count1 = data.index
    count1 = len(temp_count1)
    temp_count2 = df.index
    count2 = len(temp_count2)
    return (count1 - count2)
```

```python
count_difference(df,energy,GDP,ScimEn)
```

```python
def average_GDP(data):
    data['average GDP'] = data['2006']+data['2007']+data['2008']+data['2009']+data['2010']+data['2011']+data['2012']+data['2013']+data['2014']+data['2015']
    data['average GDP'] = data['average GDP']/10
    average_GDP = pd.Series(data['average GDP'])
    average_GDP = average_GDP.sort_values(ascending=False)
    return average_GDP
series = average_GDP(df)
```
