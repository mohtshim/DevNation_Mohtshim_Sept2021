```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

np.random.seed(0)
n = 15
x = np.linspace(0,10,n) + np.random.randn(n)/5                             
y = np.sin(x)+ x/6 + np.random.randn(n)/10

x = x.reshape(-1,1)
y = y.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(x,y,random_state = 0)
```

```python
X_train
```

# Question no 1

```python
lst = [1,3,6,9]
for elem in lst:
    poly = PolynomialFeatures(degree = elem,include_bias=True)
    X_poly = poly.fit_transform(X_train)
    #poly.fit(X_poly, y_train)
    model = LinearRegression()
    model.fit(X_poly, y_train)
    X_new = np.linspace(0, 10, 100).reshape(100, 1)
    X_new_poly = poly.transform(X_new)
    predicted_values = model.predict(X_new_poly)
    plt.scatter(X_train,y_train)
    plt.scatter(X_test,y_test,color = "yellow")
    plt.plot(X_new, predicted_values, color = 'green')
    plt.title('Polynomial Regression')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

```python
def poly_reg(x,y,deg):
    poly = PolynomialFeatures(degree = deg,include_bias=True)
    X_poly = poly.fit_transform(x)
    #poly.fit(X_poly, y_train)
    model = LinearRegression()
    model.fit(X_poly, y)
    return model
```

```python
model1 = poly_reg(X_train,y_train,1)
model2 = poly_reg(X_train,y_train,3)
model3 = poly_reg(X_train,y_train,6)
model4 = poly_reg(X_train,y_train,9)
```

```python
def visulize():
    X_new = np.linspace(0, 10, 100).reshape(100, 1)
    X_new_poly = poly.transform(X_new)
    predicted_values = model.predict(X_new_poly)
    plt.scatter(x, y, color = 'blue')
    plt.scatter(X_test,y_test,color = "yellow")
    plt.plot(X_new, model1.predict(X_new_poly), color = 'green')
    plt.plot(X_new, model2.predict(X_new_poly), color = 'red')
    plt.plot(X_new, model3.predict(X_new_poly), color = 'purple')
    plt.plot(X_new, model4.predict(X_new_poly), color = 'black')
    plt.title('Polynomial Regression with degree 1,3,6,9')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
visulize()
```

```python
def r2():
    array1=[]
    array2=[]
    for i in range(0,10):
        poly = PolynomialFeatures(degree=i)
        x_train_transform = poly.fit_transform(X_train)
        x_test_transform = poly.transform(X_test)
        lr = LinearRegression()
        lr.fit(x_train_transform, y_train)
        y_pred_test = lr.predict(x_test_transform)
        y_pred_train = lr.predict(x_train_transform)
        temp1=r2_score(y_train, y_pred_train)
        array1.append(temp1)
        temp2=r2_score(y_test, y_pred_test)
        array2.append(temp2)
    print('Tuple of two arrays',list(zip(array1, array2)))
    return list(zip(array1, array2))
```

```python
listtt= r2()
```

# Question no 3

```python
r2_score_train = []
for i in range(0,len(listtt)-1):
    r2_score_train.append(listtt[i][0])
        
def plot_r2_score(r2_score_train):
    #we want only train r2_score
    x = np.linspace(0,10,9)
    plt.plot(x,r2_score_train)
    plt.xlabel("Degree")
    plt.ylabel("r2_Score")
    plt.title("Degree vs r2_score")
    
plot_r2_score(r2_score_train)
```
