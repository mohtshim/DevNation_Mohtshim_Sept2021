```python
import pandas as pd 
pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.options.display.float_format = '{:,}'.format
df = pd.read_csv("Copy of coinmarketcap_06122017.csv")
del df["Unnamed: 0"]
df.head()
```

# Question no 4

```python
plot = df.nlargest(10,'market_cap_usd')
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 5))
# creating the bar plot
plt.bar(plot['name'], plot['market_cap_usd'])
plt.ylabel("Market capacity")
plt.xlabel("Names")
plt.title("Bar plot")
plt.show()
```

# Question no 5

```python
fig = plt.figure(figsize = (10, 5))
# creating the bar plot
plt.bar(plot['name'], plot['market_cap_usd'],log = True, color=['black', 'red','black', 'blue','blue', 'green', 'green','black', 'green', 'cyan'])
plt.ylabel("Market capacity")
plt.xlabel("Names")
plt.title("Bar plot")
plt.show()
```
